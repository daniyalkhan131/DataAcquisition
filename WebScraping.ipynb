```python
name=list()
plot=list()
twist=list()
for i in soup.find_all('div', class_="css-lmiyqy e1tmud0h8"):
  name.append(i.find('h2', class_="css-2ibyss e1tmud0h7").text.strip())  #we use find because have to find only one entry #.stripe() is to remove formatting type things
  info_list=i.find_all('p', class_="css-18vfmjb et3p2gv0")
  plot.append(info_list[0].text)
  twist.append(info_list[1].text)
```

```python
d={'name':name,'plot':plot,'twist':twist}
```

```python
df=pd.DataFrame(d)
```
