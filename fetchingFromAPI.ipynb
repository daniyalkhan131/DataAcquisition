```python
import requests
import pandas as pd
```

```python
url = "https://api.themoviedb.org/3/movie/popular?api_key=c06eb7c28b6e5db5333deb0afb2b8739&language=en-US&page=1"
requests.get(url)
```

```python
#response 200 means all goes well
```

```python
response=requests.get(url)
response.json()
```

```python
#it give back dictionary
dic=response.json()
dic['results'] #this gives list of details
```

```python
df=pd.DataFrame(dic['results'])
```

```python
df.head(3)
```

```python
df=df[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df.head()
```

```python
 #this is only for page 1, now we keep on appending datafram of every page
```

```python
df=pd.DataFrame()
```

```python
df
```

```python
for i in range(1,38887):
  response=requests.get("https://api.themoviedb.org/3/movie/popular?api_key=c06eb7c28b6e5db5333deb0afb2b8739&language=en-US&page={}".format(i))
  temp_df= pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
  df=df.append(temp_df,ignore_index=True) #idnore index is for getting 1 index of final df not like repeatinf of index as df getting added
```

```python
df.shape
```

```python
#now we can store it as csv
df.to_csv('movies.csv')
```
